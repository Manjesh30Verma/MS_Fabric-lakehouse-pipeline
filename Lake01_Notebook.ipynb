```python
# Rename columns to remove invalid characters
new_column_names = {col: col.replace(" ", "_").replace(",", "").replace(";", "").replace("(", "").replace(")", "").replace("\n", "").replace("\t", "") for col in df2.columns}
for old_col, new_col in new_column_names.items():
    df2 = df2.withColumnRenamed(old_col, new_col)

# Display the updated column names
print(df2.columns)

# Now save the DataFrame
df2.write.mode("overwrite").format("delta").saveAsTable("Job12_table")
```
